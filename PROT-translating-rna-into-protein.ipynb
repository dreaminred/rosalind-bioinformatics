```python
def transcribeRNA2DNA(string: str) -> str:
    return string.replace("U", "T")
    
```

```python
def translateRNA2Protein(rnaString):
    assert len(rnaString)%3 == 0
    dnaString = transcribeRNA2DNA(rnaString)
    proteinString = ''
    for codonIndex in range(0, len(dnaString), 3):
        codon = dnaString[codonIndex:codonIndex+3]
        proteinString += codonTab[codon]
    proteinString = proteinString.replace("*", "")
    return proteinString
```
